```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

```

#STEP 2: Load & Sort Dataset

```python
df = pd.read_csv("Smart Home Energy Consumption Optimization (1).csv")

df['timestamp'] = pd.to_datetime(df['timestamp'])
df = df.sort_values(['home_id', 'device_id', 'timestamp'])

df.head()

```

#STEP 3: Feature Engineering (Lag Feature)

```python
df['power_lag_1'] = df.groupby(['home_id', 'device_id'])['power_watt'].shift(1)
df = df.dropna()
```

#STEP 4: Select Features & Target

```python
features = [
    'power_lag_1',
    'hour_of_day',
    'day_of_week',
    'indoor_temp',
    'outdoor_temp',
    'humidity',
    'light_level',
    'price_kWh'
]

X = df[features]
y = df['power_watt']

```

#STEP 5: Time-Based Trainâ€“Test Split

```python

split = int(0.8 * len(df))

X_train, X_test = X.iloc[:split], X.iloc[split:]
y_train, y_test = y.iloc[:split], y.iloc[split:]
```

#STEP 6: Normalize X and y (0â€“1)

```python

X_scaler = MinMaxScaler()
y_scaler = MinMaxScaler()

X_train_scaled = X_scaler.fit_transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

y_train_scaled = y_scaler.fit_transform(y_train.values.reshape(-1,1))
y_test_scaled = y_scaler.transform(y_test.values.reshape(-1,1))
```

#STEP 7: Train Linear Regression (Baseline)

```python

lr = LinearRegression()
lr.fit(X_train_scaled, y_train_scaled)

```

#STEP 8: Predict (0â€“1)

```python
y_pred_lr = lr.predict(X_test_scaled)
```

#STEP 9: Evaluate Baseline (0â€“1 Metrics)

```python
baseline_mae = mean_absolute_error(y_test_scaled, y_pred_lr)
baseline_rmse = mean_squared_error(y_test_scaled, y_pred_lr, squared=False)
baseline_r2 = r2_score(y_test_scaled, y_pred_lr)

print("Baseline Linear Regression (Normalized)")
print("MAE :", baseline_mae)
print("RMSE:", baseline_rmse)
print("R2  :", baseline_r2)
```

#STEP 10: Plot Actual vs Predicted (0â€“1)

#STEP 11: Save Baseline Results (For Module 6)

```python
baseline_results = {
    "Model": "Linear Regression (Normalized)",
    "MAE": baseline_mae,
    "RMSE": baseline_rmse,
    "R2": baseline_r2
}

baseline_results
```

```python
# ===============================
# MODULE 5: LSTM MODEL DEVELOPMENT
# DEVICE-WISE ENERGY FORECASTING
# ===============================

import pandas as pd
import numpy as np
import joblib
import os
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam
```

```python
# ===============================
# CONFIGURATION
# ===============================

SEQ_LEN = 24          # last 24 hours
EPOCHS = 10
BATCH_SIZE = 512
MODEL_DIR = "model"

os.makedirs(MODEL_DIR, exist_ok=True)
```

```python
# ===============================
# LOAD DATA
# ===============================

df = pd.read_csv("Smart Home Energy Consumption Optimization (1).csv")
df['timestamp'] = pd.to_datetime(df['timestamp'])
df = df.sort_values("timestamp")
```

```python
# ===============================
# HELPER FUNCTION
# ===============================

def create_sequences(data, seq_len=24):
    X, y = [], []
    for i in range(len(data) - seq_len):
        X.append(data[i:i+seq_len])
        y.append(data[i+seq_len])
    return np.array(X), np.array(y)
```

```python
# ===============================
# DEVICE-WISE TRAINING
# ===============================

devices = df['device_id'].unique()

metrics_summary = []

for device in devices:

    print("\n===============================")
    print(f"Training model for device: {device}")
    print("===============================")
```

```python
for device in df['device_id'].unique():

    print(f"\nProcessing device: {device}")

    # -------------------------------
    # Filter device data
    # -------------------------------
    df_dev = df[df['device_id'] == device][['timestamp', 'power_watt']]
    df_dev = df_dev.sort_values('timestamp')

    values = df_dev[['power_watt']].values

    # Skip very small devices
    if len(values) < 100:
        print("Not enough data, skipping...")
        continue

    # ---- rest of device-wise model code here ----

```

```python
# -------------------------------
    # Train-test split
    # -------------------------------
split = int(0.8 * len(values))
train_values = values[:split]
test_values  = values[split:]
```

```python
# -------------------------------
    # Scaling (PER DEVICE)
    # -------------------------------
scaler = MinMaxScaler()
train_scaled = scaler.fit_transform(train_values)
test_scaled  = scaler.transform(test_values)
```

```python
 # -------------------------------
    # Create sequences
    # -------------------------------
X_train, y_train = create_sequences(train_scaled, SEQ_LEN)
X_test,  y_test  = create_sequences(test_scaled, SEQ_LEN)

X_train = X_train.reshape(X_train.shape[0], SEQ_LEN, 1)
X_test  = X_test.reshape(X_test.shape[0], SEQ_LEN, 1)
```

```python
# -------------------------------
# LSTM MODEL
# -------------------------------
model = Sequential([
    LSTM(32, input_shape=(SEQ_LEN, 1)),
    Dense(1)
])

model.compile(
    optimizer=Adam(0.001),
    loss="mse"
)
```

```python
    # -------------------------------
    # TRAIN
    # -------------------------------
model.fit(
    X_train,
    y_train,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    verbose=1
)
```

```python
 # -------------------------------
    # EVALUATION
    # -------------------------------
y_pred = model.predict(X_test, batch_size=BATCH_SIZE).flatten()

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"MAE : {mae}")
print(f"RMSE: {rmse}")
print(f"R2  : {r2}")

metrics_summary.append([
    device, mae, rmse, r2
])
```

```python
# -------------------------------
# SAVE MODEL & SCALER
# -------------------------------
model.save(f"{MODEL_DIR}/lstm_{device}.h5")
joblib.dump(scaler, f"{MODEL_DIR}/scaler_{device}.pkl")

print(f"Saved model and scaler for {device}")
```

```python
# ===============================
# MODULE 6 â€“ DEVICE-WISE LSTM MODEL
# ===============================

import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam
import joblib
import os
```

```python
# ===============================
# LOAD DATA
# ===============================

df = pd.read_csv("Smart Home Energy Consumption Optimization (1).csv")
df['timestamp'] = pd.to_datetime(df['timestamp'])
df = df.sort_values(['device_id', 'timestamp'])
```

```python
# Create folder for models if it doesn't exist
os.makedirs("models", exist_ok=True)

# ===============================
# DEVICE-WISE MODEL TRAINING & EVALUATION
# ===============================

results = []
```

```python
for device in df['device_id'].unique():
    print(f"\nProcessing device: {device}")

    # -------------------------------
    # Filter device data
    # -------------------------------
    df_dev = df[df['device_id'] == device][['timestamp', 'power_watt']]
    df_dev = df_dev.sort_values('timestamp')

    values = df_dev[['power_watt']].values

    # Skip very small devices
    if len(values) < 100:
        print("Not enough data, skipping...")
        continue
```

```python
 # -------------------------------
    # Scale data
    # -------------------------------
scaler = MinMaxScaler()
values_scaled = scaler.fit_transform(values)

    # -------------------------------
    # Create sequences
    # -------------------------------
SEQ_LEN = 24
X, y = [], []
for i in range(len(values_scaled) - SEQ_LEN):
    X.append(values_scaled[i:i+SEQ_LEN])
    y.append(values_scaled[i+SEQ_LEN])

X = np.array(X)
y = np.array(y)
    
```

```python
# -------------------------------
    # Split train/test 80/20
    # -------------------------------
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

```python
# -------------------------------
    # Build LSTM model
    # -------------------------------
model = Sequential([
    LSTM(32, return_sequences=True, input_shape=(SEQ_LEN,1)),
    LSTM(16),
    Dense(1)
])
model.compile(optimizer=Adam(0.001), loss="mse")
```

```python
# -------------------------------
    # Train model
    # -------------------------------
model.fit(
    X_train, y_train,
    epochs=15,
    batch_size=64,
    verbose=1
)
```

```python
 # -------------------------------
    # Evaluate model
    # -------------------------------
y_pred_scaled = model.predict(X_test, verbose=0)
y_pred = scaler.inverse_transform(y_pred_scaled)
y_true = scaler.inverse_transform(y_test)

mae = mean_absolute_error(y_true, y_pred)
mse = mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_true, y_pred)

print(f"{device} -> MAE: {mae:.2f}, RMSE: {rmse:.2f}, R2: {r2:.3f}")

results.append({
    "device_id": device,
    "MAE": mae,
    "RMSE": rmse,
    "R2": r2
})
```

```python
# -------------------------------
    # Save model and scaler
    # -------------------------------
model.save(f"models/{device}_lstm.h5")
joblib.dump(scaler, f"models/{device}_scaler.pkl")
print(f"Saved model and scaler for {device}")
```

```python
# ===============================
# RESULTS DATAFRAME
# ===============================
results_df = pd.DataFrame(results)
print("\nDevice-wise Model Performance:")
print(results_df)
```

```python
# ===============================
# DEVICE-WISE PREDICTION FUNCTION
# ===============================

def predict_device_energy(device_id, last_24_values):
    """
    Predict energy consumption for a device
    Inputs:
    - device_id : string, e.g., 'air_conditioner1'
    - last_24_values : list or array of last 24 readings
    Output:
    - predicted energy in Watts
    """
    # Load model and scaler
    model = load_model(f"models/{device_id}_lstm.h5", compile=False)
    scaler = joblib.load(f"models/{device_id}_scaler.pkl")

    # Prepare sequence
    seq = np.array(last_24_values).reshape(1, 24, 1)
    pred_scaled = model.predict(seq, verbose=0)
    pred = scaler.inverse_transform(pred_scaled)

    return float(pred[0][0])
```

```python
# ===============================
# FORECASTING FUNCTIONS
# ===============================

def load_device_model(device_id):
    model = load_model(f"models/{device_id}_lstm.h5", compile=False)
    scaler = joblib.load(f"models/{device_id}_scaler.pkl")
    return model, scaler
```

```python
def get_last_24_hours(device_id):
    values = df[df['device_id'] == device_id]['power_watt'].values
    if len(values) < 24:
        raise ValueError("Not enough historical data")
    return values[-24:]
```

```python
def predict_next_hour(device_id, user_input):
    model, scaler = load_device_model(device_id)

    last_23 = get_last_24_hours(device_id)[1:]
    sequence = np.append(last_23, user_input)

    seq_scaled = scaler.transform(sequence.reshape(-1,1))
    seq_scaled = seq_scaled.reshape(1,24,1)

    pred_scaled = model.predict(seq_scaled, verbose=0)
    pred = scaler.inverse_transform(pred_scaled)

    return float(pred[0][0])
```

```python
def predict_future(device_id, user_input, hours):
    model, scaler = load_device_model(device_id)

    history = list(get_last_24_hours(device_id)[1:])
    history.append(user_input)

    preds = []
    for _ in range(hours):
        seq = np.array(history[-24:])
        seq_scaled = scaler.transform(seq.reshape(-1,1))
        seq_scaled = seq_scaled.reshape(1,24,1)

        pred_scaled = model.predict(seq_scaled, verbose=0)
        pred = scaler.inverse_transform(pred_scaled)[0][0]

        preds.append(pred)
        history.append(pred)

    return np.mean(preds)
```

```python
def predict_next_week(device_id, user_input):
    return predict_future(device_id, user_input, 24*7)

def predict_next_month(device_id, user_input):
    return predict_future(device_id, user_input, 24*30)

def energy_suggestion(value):
    if value > 2000:
        return "High usage detected. Consider reducing usage."
    elif value > 1000:
        return "Moderate usage. Try optimizing consumption."
    else:
        return "Energy usage is optimal."
```

```python
# ===============================
# TEST FORECAST
# ===============================
device = "washer9"
user_input = 500

print("Next Hour:", predict_next_hour(device, user_input))
print("Next Week:", predict_next_week(device, user_input))
print("Next Month:", predict_next_month(device, user_input))
print("Suggestion:", energy_suggestion(predict_next_hour(device, user_input)))
```
