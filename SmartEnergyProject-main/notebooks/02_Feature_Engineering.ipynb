```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/House_1_cleaned_named.csv")
df.head()
```

```python
df.shape
```

```python
df['Time'] = pd.to_datetime(df['Time'])
df.set_index('Time', inplace=True)
```

```python
df.head()
```

```python
df['hour'] = df.index.hour
df['day'] = df.index.day
df['weekday'] = df.index.weekday
df['month'] = df.index.month
```

```python
df[['hour', 'day', 'weekday', 'month']].head()
```

```python
appliance_cols = [
    'Fridge', 'Freezer', 'Washing_Machine', 'Dishwasher',
    'Computer', 'Television', 'Microwave', 'Kettle', 'Toaster'
]
```

```python
df['total_appliance_load'] = df[appliance_cols].sum(axis=1)
df['mean_appliance_load'] = df[appliance_cols].mean(axis=1)
df['max_appliance_load'] = df[appliance_cols].max(axis=1)
```

```python
df[['total_appliance_load', 'mean_appliance_load', 'max_appliance_load']].head()
```

```python
df['lag1'] = df['Aggregate'].shift(1)
df['lag24'] = df['Aggregate'].shift(24)
```

```python
df['rolling_mean_3'] = df['Aggregate'].rolling(window=3).mean()
df['rolling_mean_24'] = df['Aggregate'].rolling(window=24).mean()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum().sum()
```
